```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
data = pd.read_csv("C:/Users/hp/Desktop/Everything/Work/Projects/linear regression/data sets/DailyDelhiClimateTest.csv")
df = pd.DataFrame(data)
```

```python

#my code 

x = df.iloc[:,1]
y = df.iloc[:,2]

x_data = (x - x.mean())/x.std()
y_data = (y - y.mean()) / y.std()


n = len(x_data)
w= 0 
b= 0
alpha = 0.001
epoches = 10000
indices = np.random.permutation(n)

for epoch in range (epochs):
    
    for i in indices:
        
        
        y_pred = w*x_data[i]  + b
        error = y_data[i] - y_pred
    
        gradslope = -2 *x_data[i]* error 
        gradinter = -2 * error
    
        w = w - alpha  *gradslope
        b= b - alpha * gradinter

print(w,b)  

worig = w * (y.std()/x.std())
borig = y.mean() - (worig * x.mean())

ax = df.plot.scatter(x = 'meantemp' , y ='humidity')


x_vals = np.linspace(x.min(), x.max(), 100)
yvalues = (x_vals * worig) + borig


ax.plot(x_vals, yvalues, color='red', linestyle='--', label='Trend Line')

plt.title('My DataFrame Scatter Plot with Line')
plt.legend()
plt.show()


```
