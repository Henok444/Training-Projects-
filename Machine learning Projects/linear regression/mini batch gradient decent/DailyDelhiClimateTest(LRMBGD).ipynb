```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
data = pd.read_csv("C:/Users/hp/Desktop/Everything/Work/Projects/linear regression/data sets/DailyDelhiClimateTest.csv")
df = pd.DataFrame(data)
```

```python
x= df.iloc[:,3].values
y = df.iloc[:,4].values

xb = (x-x.mean())/x.std()
yb = (y-y.mean())/y.std()


n = len(x)
batch_size = 3200
w= 0 
b= 0 
alpha = 0.01


indices = np.random.permutation(n)
"""
ch = np.array_split(indices,no_of_batch)   # this does not work 
c = len(ch)
print(ch[1])
"""
for i in range(50):
    for start in range(0 , n ,batch_size):
        end  = start + batch_size
    
        batch_index = indices[start:end]
        xi = xb[batch_index]
        yi = yb[batch_index]
    
        
       
    
        ypred = w*xi + b
        error = (yi - ypred)
        gradslope = (-2/len(xi)) * np.sum(error*xi)
        gradinter = (-2/len(xi)) * np.sum(error)
    
        w = w - (alpha * gradslope) 
        b = b - (alpha * gradinter)

print(w,b)
    



worig = w * (y.std()/x.std())
borig = y.mean() - (worig * x.mean())

ax = df.plot.scatter(x = 'vibration_x' , y ='vibration_y')


x_vals = np.linspace(x.min(), x.max(), 100)
yvalues = (x_vals * worig) + borig


ax.plot(x_vals, yvalues, color='red', linestyle='--', label='Trend Line')

plt.title('My DataFrame Scatter Plot with Line')
plt.legend()
plt.show()
```
