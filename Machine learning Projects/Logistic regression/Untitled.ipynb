```python
import numpy as np

class LogisticRegressionScratch:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.lr = learning_rate
        self.iterations = iterations
        self.theta = None

    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))

    def fit(self, X, y):
        m, n = X.shape

        # Add bias term
        X = np.c_[np.ones((m, 1)), X]

        # Initialize parameters
        self.theta = np.zeros(n + 1)

        for _ in range(self.iterations):
            z = np.dot(X, self.theta)
            h = self.sigmoid(z)

            # Gradient
            gradient = (1 / m) * np.dot(X.T, (h - y))

            # Update
            self.theta -= self.lr * gradient

    def predict_proba(self, X):
        m = X.shape[0]
        X = np.c_[np.ones((m, 1)), X]
        return self.sigmoid(np.dot(X, self.theta))

    def predict(self, X, threshold=0.5):
        probs = self.predict_proba(X)
        return (probs >= threshold).astype(int)

X1 = df[['RestingBP','Cholesterol','Age']].to_numpy()
y1 = df[['HeartDisease']].to_numpy().flatten()
X1[:,0] = (X1[:,0] - X1[:,0].mean()) / X1[:,0].std()
X1[:,1] = (X1[:,1] - X1[:,1].mean()) / X1[:,1].std()
X1[:,2] = (X1[:,2] - X1[:,2].mean()) / X1[:,2].std()

model = LogisticRegressionScratch(learning_rate=0.1, iterations=1000)
model.fit(X1, y1)

predictions = model.predict()
print(predictions)

# Generate smooth curve
X1_plot = np.linspace(0, 6, 100).reshape(-1, 1)
y1_prob = model.predict_proba(X_plot)

# Plot
plt.figure()
plt.scatter(X, y)
plt.plot(X_plot, y_prob)
plt.xlabel("X")
plt.ylabel("Probability")
plt.title("Logistic Regression Fit")
plt.show()
```
